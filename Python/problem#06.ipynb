```python
import pandas as pd 

# creating dataframe
df = pd.read_excel("pokemon.xlsx")
```

```python
df.head()
```

```python
# Pokemons whose spawn rate is less than 5%
df_2 = df[df['spawn_chance']<0.05]

print(f"Total Pokemons whose spawn rate is less than 5%: {df_2.shape[0]}")
df_2.head()
```

```python
# Pokemons that have less than 4 weaknesses
import re

def extract_words(text):
    # Use regular expression to match words inside single quotes
    pattern = r"'(\w+)'"
    matches = re.findall(pattern, text)
    
    return matches

def weaknesses_count(list):
    if len(list)<4:
        return True 
    else:
        return False
        
# converting the string into list
df['weaknesses'] = df['weaknesses'].apply(extract_words)

# Pokemons that have less than 4 weaknesses
df3 = df[df['weaknesses'].apply(weaknesses_count)]

print(f"Total  Pokemons that have less than 4 weaknesses: {df3.shape[0]}")
df3.head(5)
```

```python
# Pokemons that have no multipliers at all
df_4 = df[df['multipliers'].isna()]

print(f"Total Pokemons that have no multipliers at all: {df_4.shape[0]}")
df_4.head()
```

```python
import pandas as pd

# Convert time to seconds for each row
df['spawn_time_in_seconds'] = df['spawn_time'].fillna('00:00').apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))

# changing the column order
columns = ['id', 'num', 'name', 'img', 'type', 'height', 'weight', 'candy',
       'candy_count', 'egg', 'spawn_chance', 'avg_spawns', 'spawn_time','spawn_time_in_seconds',
       'multipliers', 'weaknesses', 'next_evolution', 'prev_evolution']

df = df[columns]
df_5 = df[df['spawn_time_in_seconds']<300]

print(f"Total Pokemons whose spawn time is less than 300 seconds : {df_5.shape[0]}")
df_5.head(5)
```
